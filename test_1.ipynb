```python
# חקירת גרפים
# 1. פילוג לקוחות לפי מגדר

By_gender =['Female', 'Male', 'Non-binary', 'Female', 'Bigender']
gender_count ={'Female': By_gender.count('Female'),
                'Male': By_gender.count('Male'),
                'Non-binary': By_gender.count('Non-binary'),
                'Bigender': By_gender.count('Bigender')}

customers.dropna(subset=['gender'], inplace=True)
plt.figure(figsize=(8, 5))
plt.bar(gender_count.keys(), gender_count.values(), color=['pink', 'red', 'blue', 'silver'])
plt.title('Table of genders')
plt.show()
```

```python
#3. היסטוגרמת פילוג לקוחות לפי גיל

plt.figure(figsize=(12,4)) 
sns.displot(data=customers, x='age', bins=30,  kde=True) 
plt.title('Customer distribution histogram by age')
plt.ylabel('Customers')
plt.show()
```
