```python
# 4. גרף המציג כמות לקוחות שהצטרפו לפי שנים

customers_df = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/main/customers.csv")
customers_df['joining_date'] = pd.to_datetime(customers_df['joining_date'], errors='coerce')
customers_df.dropna(subset=['joining_date'],inplace=True)
customers_df['joining_year'] = customers_df['joining_date'].dt.year
customers_per_year = customers_df['joining_year'].value_counts().sort_index()

plt.figure(figsize=(7, 4))
plt.bar(customers_per_year.index.astype(str), customers_per_year.values, color='#31F7F7')
plt.title('Customers who joined by year')
plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```
